#### Шаг 5. Подсчёт выручки

##### 5.1 Построение таблицы по каждому пользователю с суммарной дистанцией и длительности поездок за каждый месяц

```python
user_rides['duration'] = np.ceil(user_rides['duration']) # Округление исходного датафрейма
user_stats = pd.pivot_table(user_rides, 
                            index=['user_id', 'month'], 
                            values=['distance', 'duration', 'minute_price', 'start_ride_price', 'subscription_fee', 'subscription_type'], 
                            aggfunc={'distance':['sum', 'count'], 
                                     'duration':'sum', 
                                     'minute_price':'first',
                                     'start_ride_price':'first',
                                     'subscription_fee':'first',
                                     'subscription_type':'first'}
                            )
user_stats.columns = ['count', 'distance', 'duration', 'minute_price', 'start_ride_price', 'subscription_fee', 'subscription_type']
user_stats
```

```python
# Проверка, появились ли пропуски в сводной таблице
user_stats.isna().sum().sum()
```

##### 5.2 Добавление столбца с выручкой

```python
user_stats['month_revenue'] = user_stats['count'] * user_stats['start_ride_price'] + user_stats['duration'] * user_stats['minute_price'] + user_stats['subscription_fee']
user_stats.drop(columns=['minute_price', 'start_ride_price', 'subscription_fee'], inplace=True)
user_stats
```

##### Вывод

Сделана общая таблица расходам каждого пользователя на каждый отдельно взятый месяц (тип подписки также есть в таблице).

#### Шаг 6. Проверка гипотез

##### 6.1 Тратят ли пользователи с подпиской больше времени на поездки?

В данном случае гипотезы можно сформулировать следующим образом:

$H_0: \mu_2 = \mu_1$

$H_1: \mu_2 > \mu_1$

$\mu_2$ - Среднее время пользователей с подпиской, $\mu_1$ - среднее время пользователей без подписки 

```python
alpha = .05

results = st.ttest_ind(user_rides_sub['duration'], user_rides_free['duration'], alternative='greater')

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не отвергаем нулевую гипотезу')
```

Пользователи с подпиской действительно тратят больше времени на поездки

##### 6.2 Среднее расстояние не превышает 3130 метров?

Задачу нужно переформулировать на математический язык:

$H_0: \mu = 3130$

$H_1: \mu > 3130$

Нулевая гипотеза означает что расстояние оптимальное, а альтернативная - среднее расстояние превышает 3130 метров.

```python
optimal_value = 3130
alpha = .05

results = st.ttest_1samp(user_rides_sub['distance'], optimal_value, alternative='greater')

print('p-значение:', results.pvalue)

if (results.pvalue < alpha):
    print('Отвергаем нулевую гипотезу')
else:
    print('Не отвергаем нулевую гипотезу') 
```

Можно сказать что у самокатов есть запас по средней дистанции использования, поэтому они используются оптимально.

##### 6.3 Помесячная выручка от пользователей с подпиской больше чем от пользователей без подписки?

```python
alpha = .05

results = st.ttest_ind(user_stats.loc[user_stats['subscription_type'] == 'ultra', 'month_revenue'], 
                       user_stats.loc[user_stats['subscription_type'] == 'free', 'month_revenue'], 
                       alternative='greater')

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не отвергаем нулевую гипотезу')
```

Выручка от пользователей с подпиской действительно больше.

##### 6.4 (гипотетически) Что нужно сделать чтобы проверить, снизилось ли число обращений в техподдержку после обновления?

Поскольку выборки зависимы друг от друга - правильно будет использовать `scipy.stats.ttest_rel()`, как функцию для исследования парных выборок. В таком случае гипотезы будут выглядеть так:

$H_0: \mu_1 = \mu_0$

$H_1: \mu_1 < \mu_0$

где $\mu_0$ это усреднённое число обращений до обновления, а $\mu_1$ - после. Дальше можно делать вывод из полученного p-уровня значимости.
