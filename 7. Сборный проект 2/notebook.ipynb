### Шаг 4. Подготовка данных

Изначально нужно подготовить пайплайн для предобработки данных - категореальных признаков, категориальных упорядоченных и числовых

```python
# Порядковые категориальные фичи
ord_cols = ['level', 'workload']
cat_cols = [i for i in cat_cols if i not in ord_cols]
# Удаление целевого признака
num_cols.remove('job_satisfaction_rate')
num_cols
```

```python
# Пайплайн для предобработки данных

# One-hot encoding
ohe_pipe = Pipeline(
    [
        (
            'simpleImputer_ohe', 
            SimpleImputer(missing_values=np.nan, strategy='most_frequent')
        ),
        (
            'ohe', 
            OneHotEncoder(drop='first', handle_unknown='ignore', sparse_output=False)
        )
    ]
) 

# Ordinal encoding
ord_pipe = Pipeline(
    [
        (
            'simpleImputer_before_ord', 
            SimpleImputer(missing_values=np.nan, strategy='most_frequent')
        ),
        (
            'ord',  
            OrdinalEncoder(
                categories=[
                    ['junior', 'middle', 'sinior'], 
                    ['low', 'medium', 'high'],
                ], 
                handle_unknown='use_encoded_value', unknown_value=np.nan
            )
        ),
        (
            'simpleImputer_after_ord', 
            SimpleImputer(missing_values=np.nan, strategy='most_frequent')
        )
    ]
)

num_pipe = Pipeline(
    [
        (
            'simpleImputer_num', 
            SimpleImputer(missing_values=np.nan, strategy='mean')
        ),
        (
            'scaler', 
            MinMaxScaler()
        )
    ]
)

# Общий пайплайн
data_preprocessor = ColumnTransformer(
    [
        ('ohe', ohe_pipe, cat_cols),
        ('ord', ord_pipe, ord_cols),
        ('num', num_pipe, num_cols)
    ], 
    remainder='passthrough'
) 
```

На этом предобработка данных закончена

### Шаг 5. Обучение моделей

Нужно обучить и сравнить 2 модели, соединив модели с предобработкой. Модель линейной регрессии я буду использовать как базовую модель, поскольку она не имеет гиперпараметров, а как вторую модель буду использовать дерево решений, с подбором гиперпараметров через OptunaSearchCV

```python
pipe_final = Pipeline(
    [
        ('preprocessor', data_preprocessor),
        ('models', LinearRegression())
    ]
)
```

```python
# Сетка поиска гиперпараметров
param_distributions = [
    {
        'models': [LinearRegression(n_jobs=-1)],
    },
    {
        'models': [DecisionTreeRegressor(random_state=RANDOM_SEED)],
        'models__max_leaf_nodes': range(5, 400, 15),
        'models__min_samples_leaf': range(1, 10, 2),
        'models__min_samples_split': range(2, 10, 2),
        'models__max_depth': range(2, 150, 15)
    }
]
```

```python
# Метрика sMAPE
def smape(y_true, y_pred):
    return 100*np.mean(2*np.abs(y_pred - y_true) / (np.abs(y_pred) + np.abs(y_true)))

# Создание метрики
smape_scorer = make_scorer(smape, greater_is_better=False)
```

```python
# Создание подборщика гиперпараметров
randomized_search = RandomizedSearchCV(
    pipe_final, 
    param_distributions=param_distributions, 
    scoring=smape_scorer, 
    cv=5,
    random_state=RANDOM_SEED,
    n_jobs=-1
)
```

```python
# Разбиение обучающей выборки и поиск гиперпараметров
X_train = train.drop('job_satisfaction_rate', axis=1)
y_train = train['job_satisfaction_rate']
```

```python
randomized_search.fit(X_train, y_train)
```

```python
# Создаю общий сет, а затем делю его на таргет и остальные признаки
test = X_test.merge(y_test, 'left', on='id')
X_test = test.drop('job_satisfaction_rate', axis=1)
y_test = test['job_satisfaction_rate']
```

```python
randomized_search.score(X_test, y_test)
```

Теория: Удовлетворённость уходящих и остающихся сотрудников разная. $\mu_0$ для оставшихся сотрудников, $\mu_1$ для уходящих

$H_0$: $\mu_0 = \mu_1$

$H_1$: $\mu_0 \neq \mu_1$

```python
# Значение в .05 для alpha в данной задаче приемлемо, поскольку нет указаний по точности
alpha = .05

results = st.ttest_ind(quit['jsr'], dont_quit['jsr'], equal_var=False)

print('p-значение:', results.pvalue)

if (results.pvalue < alpha):
    print('Отвергаем нулевую гипотезу')
else:
    print('Не отвергаем нулевую гипотезу') 
```

Статистически, это также значимый признак.
